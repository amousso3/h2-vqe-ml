```python
# Code to build an example circuit with visualization

from qiskit import QuantumCircuit, QuantumRegister

# Initialize qubits, number of qubits, and their names
qubits = QuantumRegister(2, name='q')

# Create a circuit containing the qubits
circuit = QuantumCircuit(qubits)

# Initialize qubit objects using the register
q0, q1 = qubits

# Apply gates to the qubits
circuit.h(q0)
circuit.cx(q0, q1)
circuit.cx(q1,q0)
circuit.measure_all()

# Draws circuit for you
circuit.draw('mpl')
```

The following is transpilation code that will compile the above circuit into quantum computer friendly codee:

```python
service = QiskitRuntimeService()
backend = service.backend("ibm_brisbane")
pass_manager = generate_preset_pass_manager(optimzation_level=3, mode=backend)
isa_circuit = pass_manager.run(circuit)

isa_circuit.draw('mpl', idle_wires=False)
```
